```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("StudentsPerformance.csv")

print("Dataset shape (rows, columns): ", df.shape)
print("Column names: ", df.columns.tolist())
print("First 5 rows of the dataset:")
print(df.head())

print("Summary statistics:")
print(df.describe())

print("Data types and null info:")
print(df.info())
```

```python
print("Average score by columns:")
df[["math score", "reading score", "writing score"]].mean()
```

```python
print("Max score by subjects:")
df[["math score", "reading score", "writing score"]].max()
```

```python
print("Min score by subjects:")
df[["math score", "reading score", "writing score"]].min()
```

---------- Correlation ------------

Correlation is a statistical measure that shows how strongly two variables in a dataset are related. In AI and machine learning, it helps identify whether changes in one feature are associated with changes in another.

Positive correlation (+1): As one variable increases, the other also increases.
Example: More study hours → Higher exam scores.

Negative correlation (-1): As one variable increases, the other decreases.
Example: Higher product price → Lower sales.

No correlation (0): The variables have no meaningful relationship.
Example: Shoe size and exam score.

```python
# Correlation

print("Correlation:")
df[["math score", "reading score", "writing score"]].corr()
```

```python
df["total score"] = df[["math score", "reading score", "writing score"]].sum(axis=1)
df["average score"] = df["total score"] / 3
df["passed"] = np.where(df["average score"] >= 60, "Yes", "No") # condition(true) ? "Yes" : "No"

print("New columns preview (total, average, passed): ")
print(df[["total score", "average score", "passed"]].head())
```

```python
# test preparation course - based average
test_prep_group = df.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean()
print("Average Student score by test preparation:")
print(test_prep_group)

print("\n")

# gender - based average
gender_avg = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
print("Average Student score by gender:")
print(gender_avg)

print("\n")

# Students "passed" vs "failed"
pass_rate = df["passed"].value_counts(normalize=True) * 100
print("Students passed vs failed:")
print(pass_rate)
```

```python
# Convert scores to Numpy array

scores = df[["math score", "reading score", "writing score"]].values

# mean, std deviation
print("Mean:", np.mean(scores, axis=0))
print("Std deviation:", np.std(scores, axis=0))
print("\n")

# Z-scores (normalize)
z_scores = (scores - np.mean(scores, axis=0)) / np.std(scores, axis=0)
print("Z-scores:\n", z_scores[:5])

# Average is 0  | Negative -  | Positive +
```

```python
# How many students scored above 90 on average?
top_scores = df[df["average score"] >= 90]
print("Top Students who scored above 90:", len(top_scores))

# How many failed?
failed_students = df[df["passed"] == "No"]
print("Failed Students:", len(failed_students))
```
